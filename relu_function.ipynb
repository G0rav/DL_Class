```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.facecolor'] = 'white'
```

```
def relu(x):
  return (np.maximum(x,0))

def sigmoid(x):
  return (1/(1+np.exp(-x)))

x = np.linspace(-10,10,200)
y = np.linspace(-2,2,200)
```

```
plt.plot(x,np.tanh(x))
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
plt.plot(x,sigmoid(x))
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
plt.plot(x,relu(x))
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
f1 = x**3 + x**2 + x + 2

plt.plot(x,f1)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
```

```
f2 = x**2 + 5*x -10

plt.plot(x,f2)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
```

```
f3 = 10*x + 30

plt.plot(x,f3)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
```

```
o = relu(f1) - relu(f2) + relu(f3)

plt.plot(x,o)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
o = -relu(f1) + relu(f2) + relu(f3)

plt.plot(x,o)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
o = relu(f1) + relu(f2) + relu(f3)

plt.plot(x,o)
plt.show()
```
